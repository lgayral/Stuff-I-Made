```

2 + 2 + 1

```

How many different ways can one hundred be written as a sum of at least two positive integers?

On va compter `M[i,j]` le nombre de décompositions de $i$ en utilisant des entiers entre $1$ et $j$. Ultimement on veut `M[100,99]`.

````python
def Pb76():
    MAX = 100

    M = np.zeros((MAX+1,MAX+1),dtype=int)

    for i in range(1,MAX+1):
        M[i,1]=1
        M[1,i]=1

    for i in range(1,MAX+1):
        for j in range(2,i):
            M[i,j]=M[i,j-1]+M[i-j,j]
        M[i,i]=M[i,i-1]+1
        for j in range(i+1,MAX+1):
            M[i,j]=M[i,i]
            
    return M[100,99]
````

````python
from sympy import primerange

def Pb77():
    
    def count(n):
        MAX = n

        PRIME = set(primerange(1,MAX+1))

        M = [[0 for _ in range(MAX+1)] for _ in range(MAX+1)]

        for i in tqdm(range(2,MAX+1)):

            for j in range(2,i):
                M[i][j]= M[i][j-1] + ( M[i-j][j] if (j in PRIME) else 0 )

            M[i][i]=M[i][i-1] + (1 if (i in PRIME) else 0)

            for j in range(i+1,MAX+1):
                M[i][j]=M[i][i]

        for i in range(2,MAX+1):
            if M[i][i]>5000:
                return i

        return False

    count(100)
````

````python
class graph:

    def __init__(self,inE,outE):
        self.inE = inE
        self.outE = outE

    def new():
        return graph({},{})

    def add_vertex(self,v):
        if v not in self.inE:
            self.inE[v]=set()
            self.outE[v]=set()
        return None

    def transpose(self):
        self.inE,self.outE=self.outE,self.inE
        return None

    def prune_vertex(self,v):
        if v in self.inE:
            for w in self.inE[v]:
                self.outE[w].remove(v)
            for w in self.outE[v]:
                self.inE[w].remove(v)
            del self.inE[v]
            del self.outE[v]
        return None

    def vertex_list(self):
        return list(self.inE)

    def size(self):
        return len(self.inE)

    def add_edge(self,v,w):
        self.outE[v].add(w)
        self.inE[w].add(v)
        return None

    def successors(self,v):
        return self.outE[v]

    def predecessors(self,v):
        return self.inE[v]

    def in_degree(self,v):
        return len(self.inE[v])
        
    def out_degree(self,v):
        return len(self.outE[v])

    def degree(self,v):
        return out_degree(self,v)+in_degree(self,v)
````

## Problème 88

A natural number, N, that can be written as the sum and product of a given set of at least two natural numbers, $\{a_1, a_2, ... , a_k\}$ is called a product-sum number: $N = a_1 + a_2 + ... + a_k = a_1 × a_2 × ... × a_k$.

For example, $6 = 1 + 2 + 3 = 1 × 2 × 3$.

For a given set of size, k, we shall call the smallest N with this property a minimal product-sum number. The minimal product-sum numbers for sets of size, k = 2, 3, 4, 5, and 6 are as follows.

$k=2: 4 = 2 × 2 = 2 + 2$

$k=3: 6 = 1 × 2 × 3 = 1 + 2 + 3$

$k=4: 8 = 1 × 1 × 2 × 4 = 1 + 1 + 2 + 4$

$k=5: 8 = 1 × 1 × 2 × 2 × 2 = 1 + 1 + 2 + 2 + 2$

$k=6: 12 = 1 × 1 × 1 × 1 × 2 × 6 = 1 + 1 + 1 + 1 + 2 + 6$

Hence for $2≤k≤6$, the sum of all the minimal product-sum numbers is $4+6+8+12 = 30$; note that 8 is only counted once in the sum.

In fact, as the complete set of minimal product-sum numbers for $2≤k≤12$ is $\{4, 6, 8, 12, 15, 16\}$, the sum is 61.

What is the sum of all the minimal product-sum numbers for $2≤k≤12000$?

````python
def Pb88(n=12000):

    def FACTORLIST(n):
        L=[[] for _ in range(n+1)]
        L[1].append([])
        for i in range(2,n+1):    
            for j in range(1,n//i + 1):
                p=i*j
                for F in L[j]:
                    L[p].append(F+[i])
        return L
    
    L=FACTORLIST(2*n+1)
    
    MINPRODSUM=[0 for _ in range(n+1)]
    
    for k in range(2,2*n+1):
        for F in L[k]:
            no = k - sum(F)
            psl=len(F)+no
            if psl<=n:
                if MINPRODSUM[psl]:
                    MINPRODSUM[psl]=min(MINPRODSUM[psl],k)
                else:
                    MINPRODSUM[psl]=k
    
    return sum(set(MINPRODSUM[2:n+1]))
````

```

{1, 2, 3, 4, 5, 6} is distinct from {1, 2, 3, 4, 5, 9}
```

## Problème 95

The proper divisors of a number are all the divisors excluding the number itself. For example, the proper divisors of 28 are 1, 2, 4, 7, and 14. As the sum of these divisors is equal to 28, we call it a perfect number.

Interestingly the sum of the proper divisors of 220 is 284 and the sum of the proper divisors of 284 is 220, forming a chain of two numbers. For this reason, 220 and 284 are called an amicable pair.

Perhaps less well known are longer chains. For example, starting with 12496, we form a chain of five numbers:

12496 → 14288 → 15472 → 14536 → 14264 (→ 12496 → ...)

Since this chain returns to its starting point, it is called an amicable chain.

Find the smallest member of the longest amicable chain with no element exceeding one million.


````python
from tqdm.notebook import tqdm_notebook as tqdm

def PROPDIVLIST(n):
    D=[[1] for _ in range(n+1)]
    D[0]=D[1]=[]
    
    for i in tqdm(range(2,n+1)):
            for k in range(2,n//i+1):
                D[k*i].append(i)
    return D

def Pb95(N=10**6):
    
    D=PROPDIVLIST(N)
   
    def iterate(n):
        return sum(D[n])
    
    SMALLESTCHAIN={}
    CHAINLENGTH={}
    
    smallest = 0
    maxlength = 0

    for i in tqdm(range(N+1)):
        L=[]
        head=i
        while (head not in SMALLESTCHAIN) and (head<=N) and (head not in L):
            L.append(head)
            head=iterate(head)

        if head>N:
            for j in L:
                SMALLESTCHAIN[j]=-1
                CHAINLENGTH[j]=-1
        elif head in SMALLESTCHAIN:
            for j in L:
                SMALLESTCHAIN[j]=SMALLESTCHAIN[head]
                CHAINLENGTH[j]=CHAINLENGTH[head]
        else:
            ind=L.index(head)
            CL=len(L)-ind
            SC=min(L[ind:])
            for j in L:
                SMALLESTCHAIN[j]=SC
                CHAINLENGTH[j]=CL
                
            if CL > maxlength:
                smallest=SC
                maxlength=CL
            elif CL==maxlength and SC<smallest:
                smallest=SC

    return smallest
````

## Problème 121

Tour n: proba 1/(n+1) d'avoir un bleu.

The player pays £1 to play and wins if they have taken more blue discs than red discs at the end of the game.

If the game is played for four turns, the probability of a player winning is exactly 11/120, and so the maximum prize fund the banker should allocate for winning in this game would be £10 before they would expect to incur a loss. Note that any payout will be a whole number of pounds and also includes the original £1 paid to play the game, so in the example given the player actually wins £9.

Find the maximum prize fund that should be allocated to a single game in which fifteen turns are played.

````python
from fractions import Fraction

def pwin_aux(L,k):
    if k>len(L):
        return Fraction(0)
    elif k==0:
        return Fraction(1)
    else:
        return L[0]*pwin_aux(L[1:],k-1) + (1-L[0])*pwin_aux(L[1:],k)
    
def pwin_comp(turns):
    L=[Fraction(1,n) for n in range(2,turns+2)]
    return pwin_aux(L,(turns+2)//2)

def Pb121(turns):
    pwin = pwin_comp(turns)
    return int(1/pwin)
````

````python
import numpy as np
from math import isqrt

def Pb125(MAXVAL=10**8):

    SIZE=isqrt(MAXVAL+1)+1

    L=np.zeros((SIZE,SIZE),dtype=np.int64)
    PAL=np.zeros((SIZE,SIZE),dtype=bool)
    OVERFLOW=np.zeros((SIZE,SIZE),dtype=bool)
    
    for i in range(SIZE):
        L[i,i]=(i+1)**2
    for d in range(1,SIZE):
        for i in range(SIZE-d):
            l = L[i,i+d-1]+L[i+d,i+d]
            if OVERFLOW[i,i+d-1] or l>MAXVAL:
                OVERFLOW[i,i+d] = True
                break
            L[i,i+d]=l
            PAL[i,i+d] = (str(l)==str(l)[::-1])
    return sum(set(L[PAL]))
````
